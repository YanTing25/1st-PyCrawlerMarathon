### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import re
import time

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

```python
'''
Your Code
'''
today=time.strftime("%Y/%m/%d", time.localtime())
regex=today
pattern=re.compile(regex)
print('===='+today+'====')
l=list()
for i in soup.find(class_="part_list_2").find_all('h3'):
    date=i.find(class_="date").text
    result=re.match(pattern,date)
    if result:
        print(i.em.text+'\t'+i.a.text)
        l.append([i.em.text,i.a.text])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
import itertools
keys=lambda x:x[0]
l.sort()
ans=list()
for key,group in itertools.groupby(l, keys):
    print (key + ": " + str(list(group)))
    ans.append(key)
    ans.append(list(group))
    print('---------------------------------------------------------')
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
max=['',1]
for i in ans:
    print(i)
    if len(i)>max[1]:
        max[0]=i[0][0]
        max[1]=len(i)
print('最多篇: '+max[0]+'類 , 共 '+str(max[1])+' 篇')
```
