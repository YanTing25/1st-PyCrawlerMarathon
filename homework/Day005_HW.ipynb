## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
import json

r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=json.loads(r.text)
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''

print("資料筆數:"+str(len(response))+"\n欄位:")
for i in response[0]:
    print("\t"+i)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
for data in response:
    print("標題:"+data['title'])
    print("-貼文時間:"+data['createdAt'])
    print("-留言人數:"+str(data['commentCount']))
    print("-按讚人數:"+str(data['likeCount']))
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''
#熱門文章
comment=0
like=0
for data in response:
    comment+=data['commentCount']
    like+=data['likeCount']
print("熱門文章平均留言數:"+str(comment/len(response)))
print("熱門文章平均按讚數:"+str(like/len(response)))

#非熱門文章
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response=json.loads(r.text)
comment=0
like=0
for data in response:
    comment+=data['commentCount']
    like+=data['likeCount']
print("非熱門文章平均留言數:"+str(comment/len(response)))
print("非熱門文章平均按讚數:"+str(like/len(response)))
```
