```python
import requests
import json

r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=json.loads(r.text)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

'''
Your Code
'''
for data in response:
    print("æ¨™é¡Œ:"+data['title'])
    print("-è²¼æ–‡æ™‚é–“:"+data['createdAt'])
    print("-ç•™è¨€äººæ•¸:"+str(data['commentCount']))
    print("-æŒ‰è®šäººæ•¸:"+str(data['likeCount']))
```
